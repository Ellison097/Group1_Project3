```python
import pandas as pd
import glob
import os


def count_total_rows():
    # 定义输出目录
    output_dir = "Project3_Data_Clean_step1"

    # 查找所有clean2文件
    clean2_files = glob.glob(os.path.join(output_dir, "*_clean2.csv"))

    # 初始化总行数
    total_rows = 0

    # 打印每个文件的行数并累加
    print("各文件行数统计:")
    for file in sorted(clean2_files):
        df = pd.read_csv(file)
        rows = len(df)
        total_rows += rows
        print(f"{os.path.basename(file)}: {rows} 行")

    print(f"\n所有clean2文件总计: {total_rows} 行")

    return total_rows


if __name__ == "__main__":
    count_total_rows()
```

```python
import requests


def query_crossref(title):
    url = "https://api.crossref.org/works"
    params = {"query.title": title, "select": "DOI,title,author,issued,publisher"}
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json().get("message", {}).get("items", [])
    return []
```

```python
from habanero import Crossref


def get_biblio_from_title(title):
    cr = Crossref()
    # Search works by title
    results = cr.works(query_title=title, limit=1)
    if results["message"]["items"]:
        item = results["message"]["items"][0]
        biblio = {
            "OutputTitle": item.get("title", [""])[0],
            "OutputType": item.get("type", ""),
            "OutputStatus": item.get("status", ""),
            "OutputVenue": item.get("container-title", [""])[0],
            "OutputYear": item.get("issued", {}).get("date-parts", [[None]])[0][0],
            "OutputMonth": item.get("issued", {}).get("date-parts", [[None, None]])[0][1]
            if len(item.get("issued", {}).get("date-parts", [[None]]))[0] > 1
            else None,
            "OutputVolume": item.get("volume", ""),
            "OutputNumber": item.get("issue", ""),
            "OutputPages": item.get("page", ""),
            "OutputBiblio": item.get("DOI", ""),
            "OutputPI": ", ".join([a.get("family", "") for a in item.get("author", [])]) if "author" in item else "",
        }
        return biblio
    else:
        return None


# Example usage
title = "Importers, Exporters, and Multinationals: A Portrait of Firms in the U.S. that Trade Goods"
biblio_info = get_biblio_from_title(title)
print(biblio_info)
```

```python
import asyncio
import aiohttp
import pandas as pd
from fuzzywuzzy import fuzz
import json


async def test_crossref_api(title):
    """Test what the Crossref API returns for a given title."""
    print(f"\n\nTesting title: '{title}'")

    # Clean the title for display
    clean_title = title.lower().strip()
    print(f"Cleaned title: '{clean_title}'")

    # Prepare the query URL
    query = f"query.title={title.replace(' ', '+')}"
    url = f"https://api.crossref.org/works?{query}&rows=5"
    print(f"API URL: {url}")

    async with aiohttp.ClientSession() as session:
        try:
            async with session.get(url) as response:
                print(f"Response status: {response.status}")

                if response.status != 200:
                    print("Error: Non-200 response")
                    return None

                data = await response.json()

                if "message" not in data or "items" not in data["message"] or not data["message"]["items"]:
                    print("No items found in response")
                    return None

                # Get the items returned by the search
                items = data["message"]["items"]
                print(f"Number of items returned: {len(items)}")

                # Print the first item's title for comparison
                if items:
                    for i, item in enumerate(items[:3]):  # Show top 3 results
                        if "title" in item and item["title"]:
                            item_title = item["title"][0] if isinstance(item["title"], list) else item["title"]
                            print(f"\nResult #{i+1}:")
                            print(f"Title: {item_title}")

                            # Calculate similarity score
                            score = fuzz.ratio(clean_title, item_title.lower())
                            print(f"Similarity score: {score}")

                            # Print some key fields
                            print(f"Type: {item.get('type', 'N/A')}")
                            print(f"DOI: {item.get('DOI', 'N/A')}")

                            # Print container title if available
                            container = item.get("container-title", ["N/A"])
                            container = container[0] if isinstance(container, list) and container else "N/A"
                            print(f"Container: {container}")

                return items

        except Exception as e:
            print(f"Error: {e}")
            return None


async def main():
    # Load just 3 rows from the dataset
    df = pd.read_csv("Project3_Data_Clean_step1/group3_clean2.csv", nrows=3)
    print(f"Loaded {len(df)} rows from dataset")
    print(f"Columns: {df.columns.tolist()}")

    # Print the titles
    print("\nTitles in dataset:")
    for i, title in enumerate(df["Title"]):
        print(f"{i+1}. {title}")

    # Test each title
    for title in df["Title"]:
        await test_crossref_api(title)

    # Also test with a known good title for comparison
    print("\n\n--- Testing with a known good title ---")
    await test_crossref_api(
        "The China Syndrome: Local Labor Market Effects of Import Competition in the United States"
    )


if __name__ == "__main__":
    asyncio.run(main())
```
